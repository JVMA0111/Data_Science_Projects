#Análise exploratória

```python
import pandas as pd
dados = pd.read_csv('Netflix_Engagement_Report_2023_Jan-Jun_cleaned.csv')
dados
```

```python
dados['Release Date'] = pd.to_datetime(dados['Release Date'])
dados.sort_values('Release Date', ascending = False)
```

```python
hours_viewed = dados.drop(columns = ['Available Globally', 'Release Date'])
hours_viewed
```

```python
release_dates = dados.drop(columns = ['Available Globally', 'Hours Viewed'])
release_dates
```

###Top 10 filmes por tempo assistido

```python
top10 = hours_viewed.head(10)
top10
```

###Distribuição percentual

O valor representado pelo top 10 não se destaca grandemente ao resto do valor amostral, visto isso, a análise aponta para uma boa heterogeneidade das horas assistidas. Ou seja, não há uma diferença significativa em popularidade para os programas mais assistidos quando comparado ao montante restante.

```python
total = hours_viewed['Hours Viewed'].sum()
hours_viewed['Percentual (%)'] = hours_viewed['Hours Viewed']/total * 100
print("O valor percentual das horas vistas representado pelos 10 programas mais assistidos é de: ", (hours_viewed.head(10))['Percentual (%)'].sum())
```

#Visualização dos dados

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data = top10, x = 'Hours Viewed', y = 'Title', palette = 'viridis')
plt.xlabel('Hours Viewed(in bilions)')
plt.title('Top 10 most viewed shows on netflix(jan-jun 2023)', fontsize = 16)
```

```python
total = top10['Hours Viewed'].sum()
top10['Percentual (%)'] = top10['Hours Viewed']/total * 100
top10
```

###Distribuição percentual dentre os shows mais assistidos

######Na análise do gráfico podemos perceber um destaque para The Night Agent, Ginny & Georgia, The Glory.

```python
plt.pie(top10['Percentual (%)'], labels=top10['Title'], autopct='%1.1f%%')
plt.figure(figsize = [20,10])
```
