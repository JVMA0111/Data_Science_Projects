#Nova coluna com a classificação de acordo com o tema

```python
classe_lista = []
for CDU in emprestimos_convertido['localizacao']:
  if(CDU < 100):
    classe_lista.append('Generalidades')
  elif(CDU < 200):
    classe_lista.append('Filosofia e psicologia')
  elif(CDU < 300):
    classe_lista.append('Religião')
  elif(CDU < 400):
    classe_lista.append('Ciências sociais')
  elif(CDU < 500):
    classe_lista.append('Classe vaga')
  elif(CDU < 600):
    classe_lista.append('Matemática e ciências naturais')
  elif(CDU < 700):
    classe_lista.append('Ciências aplicadas')
  elif(CDU < 800):
    classe_lista.append('Belas artes')
  elif(CDU < 900):
    classe_lista.append('Linguagem')
  else:
    classe_lista.append('Geografia. Biografia. História.')

emprestimos_convertido['classe'] = classe_lista
emprestimos_convertido.head()
```

#Dia 3 Exploração de dados

Objetivo: Descobrir se a quantidade de empréstimos está diminuindo, aumentando ou permanecendo igual. Gerenciar a programação de férias dos colaboradores de acordo com os meses de menor demanda. Entender quais meses possuem a maior quantidade de empréstimos realizados. Entender quais são os horários com a maior quantidade de empréstimos ao longo do dia.

#Separando a quantidade de empréstimos por ano

```python

emprestimos_convertido.data_emprestimo = pd.to_datetime(emprestimos_convertido.data_emprestimo)
quantidade_de_emprestimos = emprestimos_convertido.copy().iloc[:,[6,2]]
quantidade_de_emprestimos.data_emprestimo = quantidade_de_emprestimos.data_emprestimo.dt.year
quantidade_de_emprestimos_ano = quantidade_de_emprestimos.value_counts().groupby(["data_emprestimo"]).sum().sort_values(ascending=False).to_frame()
quantidade_de_emprestimos_ano.index.rename('Ano',inplace=True)
quantidade_de_emprestimos_ano.columns = ['Quantidade de empréstimos']
quantidade_de_emprestimos_ano
```

#Plotagem em gráfico de linhas
Em uma primeira análise visual, já conseguimos observar uma queda desde 2013, com representação mais busca entre 2019 e 2020.

```python
import seaborn as sns
grafico_de_linha = sns.lineplot(data = quantidade_de_emprestimos_ano, x = quantidade_de_emprestimos_ano.index, y = "Quantidade de empréstimos")
grafico_de_linha.set_title("Quantidade de empréstimos por ano")
```

#Separando a quantidade de empréstimos por hora

```python
emprestimos_convertido.data_emprestimo = pd.to_datetime(emprestimos_convertido.data_emprestimo)
quantidade_de_emprestimos = emprestimos_convertido.copy().iloc[:,[0,2]]
quantidade_de_emprestimos.data_emprestimo = quantidade_de_emprestimos.data_emprestimo.dt.hour
quantidade_de_emprestimos_hora = quantidade_de_emprestimos.value_counts().groupby(["data_emprestimo"]).sum().sort_values(ascending=False).to_frame()
quantidade_de_emprestimos_hora.index.rename('Horário',inplace=True)
quantidade_de_emprestimos_hora.columns = ["Quantidade de empréstimos"]
quantidade_de_emprestimos_hora
```

#Gráfico de barras da distribuição dos empréstimos por horários do dia
A análise dos dados indica que os períodos entre 10-12, 14-18 são mais demandados.

```python
grafico_de_barras = sns.barplot(data = quantidade_de_emprestimos_hora, x = quantidade_de_emprestimos_hora.index, y = "Quantidade de empréstimos", palette = 'flare')
grafico_de_barras.set_title("Quantidade de empréstimos por hora")
```

#Separando a quantidade de empréstimos por mês
Conseguimos analisar e aferir que os meses de maior demanda são março e agosto. E os de menor demanda são dezembro e janeiro, sendo assim provavelmente melhores para as férias dos funcionários.

```python
emprestimos_convertido.data_emprestimo = pd.to_datetime(emprestimos_convertido.data_emprestimo)

quantidade_de_emprestimos = emprestimos_convertido.copy().iloc[:,[0,2]]
quantidade_de_emprestimos.data_emprestimo = quantidade_de_emprestimos.data_emprestimo.dt.month
quantidade_de_emprestimos_mes = quantidade_de_emprestimos.value_counts().groupby(["data_emprestimo"]).sum().to_frame()
quantidade_de_emprestimos_mes.index.rename('Mês',inplace=True)
quantidade_de_emprestimos_mes.columns = ["Quantidade de empréstimos"]

meses = { 1: 'Janeiro',
          2: 'Fevereiro',
          3 : 'Março',
          4 : 'Abril',
          5 : 'Maio',
          6 : 'Junho',
          7 : 'Julho',
          8 : 'Agosto',
          9 : 'Setembro',
          10 : 'Outubro',
          11 : 'Novembro',
          12 : 'Dezembro' }
quantidade_de_emprestimos_mes.index = quantidade_de_emprestimos_mes.index.map(meses)
quantidade_de_emprestimos_mes
```

#Gráfico de linhas da quantidade de empréstimos por mês

```python
import matplotlib.ticker as ticker
grafico_de_linha_mes = sns.lineplot(data = quantidade_de_emprestimos_mes, x = quantidade_de_emprestimos_mes.index, y = "Quantidade de empréstimos")
grafico_de_linha_mes.set_title("Quantidade de empréstimos por mês")
grafico_de_linha_mes.tick_params(axis='x',rotation=50)
```

#Extraindo dados do tipo de vínculo e adicionando a porcentagem

```python

def porcentagem_de_emprestimos(df):
    quantidade_de_emprestimos_total = df["Quantidade de empréstimos"].sum()

    df["Porcentagem"] = (df["Quantidade de empréstimos"] / quantidade_de_emprestimos_total) * 100

    df["Porcentagem"] = df["Porcentagem"].round(2)

    return df
```

```python
quantidade_de_emprestimos = emprestimos_convertido.copy().iloc[:,[5,6]]
quantidade_de_emprestimos_vinculo = quantidade_de_emprestimos.value_counts().groupby(["tipo_vinculo_usuario"]).sum().sort_values(ascending=False).to_frame()
quantidade_de_emprestimos_vinculo.index.rename('Tipo de Vínculo',inplace=True)
quantidade_de_emprestimos_vinculo.columns = ['Quantidade de empréstimos']
porcentagem_de_emprestimos(quantidade_de_emprestimos_vinculo)
quantidade_de_emprestimos_vinculo
```

O público que mais utiliza a biblioteca é representado pelos alunos de graduação. Isso possivelmente acontece por se tratarem de uma porcentagem maior da população de estudantes. Tendo esse dado em vista pode ser que seja interessante a disponibilização de livros de interesse para esse perfil de estudantes.

```python
def diferenca_percentual_ano_anterior(x,y):
  return round(((x / y * 100) - 100),2)
```
