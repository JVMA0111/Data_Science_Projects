#Primeira análise exploratória e de visualização de dados
Estou iniciando meus estudos no campo de análise de dados e esse foi meu primeiro relatório utilizando google colab, linguagem python, e algumas bibliotecas como seaborn, pandas e matplotlib.

```python
import pandas as pd
diabetes = pd.read_csv("diabetes.csv")
diabetes.columns = ["Gravidez", "Glicose", "Pressao Sanguinea",	"Espessura da Pele",	"Insulina",	"IMC",	"DiabetesPedigreeFunction",	"Idade",	"Desfecho"]
diabeticos = diabetes.query("Desfecho == 1").Desfecho
print("A quantidade de pacientes com diabetes foi:", diabeticos.count())
print("A quantidade total de pacientes foi:", diabetes.count().Gravidez)
```

```python
diabetes.head()
```

#A glicose possui correlação positiva com o desenvolvimento de diabetes?

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlacao_glicose = diabetes.groupby("Desfecho").Glicose.mean().reset_index()
sns.regplot(data=diabetes, x="Glicose", y="Desfecho", ci=None, line_kws={'color':'red'})
sns.scatterplot(data=correlacao_glicose, x="Glicose", y="Desfecho", color='black', marker='x', s=200)
plt.show()
```
